```python
import numpy as np
import matplotlib.pyplot as plt
mean1 = np.array([0,0])
mean2 = np.array([1,1])
mean3 = np.array([1,0])
mean4 = np.array([0,1])
t = 0.08
sigma = np.array([[t**2,0],[0,t**2]])
point_num = 90
```

```python
m1_class = np.random.multivariate_normal(mean1,sigma,point_num)
m2_class = np.random.multivariate_normal(mean2,sigma,point_num)
m3_class = np.random.multivariate_normal(mean3,sigma,point_num)
m4_class = np.random.multivariate_normal(mean4,sigma,point_num)
plt.scatter(m1_class[0:,0],m1_class[0:,1],color = "blue")
plt.scatter(m2_class[0:,0],m2_class[0:,1],color = "blue")
plt.scatter(m3_class[0:,0],m3_class[0:,1],color = "red")
plt.scatter(m4_class[0:,0],m4_class[0:,1],color = "red")
plt.show()
```

```python
c1= np.vstack([m1_class,m2_class])
c2= np.vstack([m3_class,m4_class])
data = np.vstack([c1,c2])
```

```python
lable_0 = np.zeros(180)
lable_1 = np.ones(180)
lable = np.hstack([lable_0,lable_1])
lable.shape
```

```python
from sklearn.linear_model import Perceptron
ppn = Perceptron(max_iter=40,eta0=0.1)
ppn.fit(data,lable)
```

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(units=10,input_dim = 2,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))

model.compile(loss='mse',optimizer='sgd',metrics=['accuracy'])
model.summary()
model.fit(data,lable,epochs=500,batch_size=10)
```

```python
from matplotlib.colors import ListedColormap


def plot_decision_regions(X, y, classifier, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                         np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # plot class samples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                    alpha=0.8, c=cmap(idx),
                    marker=markers[idx], label=cl)
```

```python
plot_decision_regions(data, lable, classifier=model)
```
